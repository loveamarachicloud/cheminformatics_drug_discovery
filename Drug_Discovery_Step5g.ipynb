# Step 5g: Comparing Fingerprints of Different Molecules
Explanation Script: "Now let's compare which bits are active in chloroquine versus ibuprofen. This shows us structural similarities and differences."

```python
from rdkit import Chem
from rdkit.Chem import AllChem, Draw
import numpy as np
import matplotlib.pyplot as plt


chloroquine = Chem.MolFromSmiles ("CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl")
ibuprofen = Chem.MolFromSmiles("CC(C)CC1=CC=C(C=C1)C(C)C(=O)O")

# Generate fingerprints
bit_info_chl = {}
bit_info_ibu = {}
fp_chl = AllChem.GetMorganFingerprintAsBitVect(chloroquine, radius=2, nBits=128, bitInfo=bit_info_chl)
fp_ibu = AllChem.GetMorganFingerprintAsBitVect(ibuprofen, radius=2, nBits=128, bitInfo=bit_info_ibu)

# Convert to arrays
chl_array = np.array(list(fp_chl))
ibu_array = np.array(list(fp_ibu))

print("First 30 bits comparison:")
print("Bit  Chloroquine Ibuprofen")
print("-" * 25)
for i in range(30):
    print(f"{i:2}    {chl_array[i]}        {ibu_array[i]}")

# Find common bits
common_bits = [i for i in range(128) if chl_array[i] == 1 and ibu_array[i] == 1]
chl_only = [i for i in range(128) if chl_array[i] == 1 and ibu_array[i] == 0]
ibu_only = [i for i in range(128) if chl_array[i] == 0 and ibu_array[i] == 1]

print(f"\nCommon bits: {common_bits[:10]}...")
print(f"Total common: {len(common_bits)}")
print(f"Chloroquine -only bits: {len(chl_only)}")
print(f"Ibuprofen-only bits: {len(ibu_only)}")

```

"Both molecules share 10 common fingerprint bits – these correspond to structural fragments that appear in both chloroquine and ibuprofen. These shared bits likely represent general chemical environments such as aromatic ring structures and carbon chain fragments that exist in both molecules.

Chloroquine has 30 unique bits that are not present in ibuprofen. These bits represent structural features characteristic of chloroquine, including the quinoline aromatic ring system, the chlorine substitution on the aromatic ring, and the nitrogen-containing side chain with tertiary amine groups.

Ibuprofen has 14 unique bits that are not present in chloroquine. These unique bits correspond to structural fragments specific to ibuprofen, such as its carboxylic acid functional group and the branched isobutyl side chain attached to the aromatic ring.

Morgan fingerprints capture these structural patterns by encoding atom environments around each atom within a defined radius. When two molecules share fingerprint bits, it means they contain similar local chemical environments.

Machine learning models use these fingerprint patterns as numerical features, allowing them to compare molecules and learn which structural fragments are associated with biological activity or pharmacological properties."
