# Step 3: Visualizing Multiple Molecules at Once
Explanation Script: "Let's visualize several molecules in a grid. This is great for comparing structures. We'll create a list of SMILES for 10 drugs and display them side by side."

```python
from rdkit import Chem
from rdkit.Chem import Draw

# List of SMILES (urea,metformin, chloroquine isoniazid pyrimethamine, salbutamol,aspirin,caffeine, ibuprofen, propofol)
smiles_list  = [ "C(=O)(N)N", # urea
"CN(C)C(=N)N=C(N)N", # metformin
"CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl", # chloroquine
"C1=CN=CC=C1C(=O)NN", # isoniazid
"CCC1=C(C(=NC(=N1)N)N)C2=CC=C(C=C2)Cl", # pyrimethamine
"CC(C)(C)NCC(C1=CC(=C(C=C1)O)CO)O", # salbutamol
 "CC(=O)OC1=CC=CC=C1C(=O)O",  # aspirin
"CN1C=NC2=C1C(=O)N(C(=O)N2C)C",  # caffeine
"CC(C)CC1=CC=C(C=C1)C(C)C(=O)O",  # ibuprofen
"CC(C)C1=C(C(=CC=C1)C(C)C)O", # propofol 
]

# Convert to molecule objects
mols = [Chem.MolFromSmiles(s) for s in smiles_list]

# Draw grid
img = Draw.MolsToGridImage(mols, molsPerRow=2, subImgSize=(250,250))
img
```
