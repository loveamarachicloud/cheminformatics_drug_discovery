# Step 5b: Comparing Molecular Fingerprints
Explanation Script: "Let's compare fingerprints of different molecules to understand similarity. Similar molecules should have more bits in common."

```python
from rdkit import Chem
from rdkit.Chem import AllChem, DataStructs

chloroquine = Chem.MolFromSmiles("CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl")
pyrimethamine = Chem.MolFromSmiles("CCC1=C(C(=NC(=N1)N)N)C2=CC=C(C=C2)Cl")
benzene = Chem.MolFromSmiles("c1ccccc1")

fp_chl = AllChem.GetMorganFingerprintAsBitVect(chloroquine, radius=2, nBits=1024)
fp_pyr = AllChem.GetMorganFingerprintAsBitVect(pyrimethamine, radius=2, nBits=1024)
fp_ben = AllChem.GetMorganFingerprintAsBitVect(benzene, radius=2, nBits=1024)

sim1 = DataStructs.TanimotoSimilarity(fp_chl, fp_pyr)
sim2 = DataStructs.TanimotoSimilarity(fp_chl, fp_ben)

print(f" Chloroquine vs pyrimethamine: {sim1:.2%}")
print(f" Chloroquine vs Benzene: {sim2:.2%}")

```

```python
These results show that chloroquine andpyrimethamine share more structural features, which is expected because both are antimalarial drugs with more complex chemical structures. Benzene is a simple aromatic ring, so it shares very few structural features with chloroquine.
This demonstrates how molecular fingerprints andTanimoto similarity can be used to quantify structural similarity between chemical compounds, a key concept in cheminformatics and drug discovery.
```
