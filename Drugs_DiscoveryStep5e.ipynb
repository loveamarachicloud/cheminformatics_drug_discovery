# Step 5e: What Substructure Does Each Bit Represent?
Explanation Script: "RDKit can tell us exactly which atom environments create each '1' in the fingerprint. Let's map bit 2 and bit 5 to actual molecular fragments."

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import SimilarityMaps
import matplotlib.pyplot as plt

# Chloroquine 
smiles = "CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl"
mol = Chem.MolFromSmiles(smiles)

# Generate fingerprint with bit info
fp = AllChem.GetMorganFingerprint(mol, radius=2)
bit_info = {}

# Get bit 2 details
bit_2_fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=2, nBits=128, bitInfo=bit_info)
print("Bit 2 appears in these atom environments:")
if 2 in bit_info:
    for atom, radius in bit_info[2]:
        print(f"  - Centered on atom {atom} with radius {radius}")
        
        # Get the environment
        env = Chem.FindAtomEnvironmentOfRadiusN(mol, radius, atom)
        amap = {}
        submol = Chem.PathToSubmol(mol, env, atomMap=amap)
        print(f"    Substructure atoms: {sorted(amap.keys())}")

# Get bit 5 details
print("\nBit 5 appears in these atom environments:")
if 5 in bit_info:
    for atom, radius in bit_info[5]:
        print(f"  - Centered on atom {atom} with radius {radius}")


```


### Interpreting Morgan Fingerprint Bits for Chloroquine
This result shows that bit 2 = 1 because chloroquine contains an atom environment centered on atom 10 with radius 2.  
This environment includes atoms [7, 8, 9, 10, 11, 12, 20], which form part of the quinoline ring system and its surrounding chemical structure.
Morgan fingerprints encode molecular structure by examining the local chemical environment around each atom.  
For each atom, RDKit looks at neighboring atoms within a specified radius and converts that environment into a hashed fingerprint bit.
In this example, the fingerprint detected a structural environment characteristic of the aromatic quinoline ring and nearby atoms, which activates bit 2.
Each fingerprint bit does not correspond to a single fixed fragment. Instead, a bit becomes 1 whenever any atom environment in the molecule matches the pattern associated with that bit.
Therefore, molecular fingerprints capture structural information from multiple parts of the molecule, including:
Aromatic ring systems  
Nitrogen-containing functional groups  
Side-chain environments  
Substituted aromatic fragments  
These encoded structural patterns allow machine learning models to compare molecules and detect features that correlate with biological activity, such as the antimalarial activity of chloroquine.
