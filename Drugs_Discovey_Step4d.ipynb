# Step 4d: LogP - Third Lipinski Rule
Explanation Script: "The third rule involves LogP, which measures how fat-soluble a molecule is. LogP should be less than 5. Higher LogP means the drug prefers fatty environments, which affects absorption."

```python
from rdkit import Chem
from rdkit.Chem import Descriptors
# Chloroquine 
chloroquine_smiles = "CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl"
mol_chl = Chem.MolFromSmiles(chloroquine_smiles)
logp_chl = Descriptors.MolLogP(mol_chl)
print(f"chloroquine LogP: {logp_chl:.2f}")

# Metformin 
metformin_smiles = "CN(C)C(=N)N=C(N)N"
mol_met = Chem.MolFromSmiles(metformin_smiles)
logp_met = Descriptors.MolLogP(mol_met)
print(f"metformin LogP: {logp_met:.2f}")

# Ibuprofen
ibuprofen_smiles = "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"
mol_ibu = Chem.MolFromSmiles(ibuprofen_smiles)
logp_ibu = Descriptors.MolLogP(mol_ibu)
print(f"Ibuprofen LogP: {logp_ibu:.2f}")
# Pyrimethamine
pyrimethamine_smiles = "CCC1=C(C(=NC(=N1)N)N)C2=CC=C(C=C2)Cl"
mol_pyr= Chem.MolFromSmiles(pyrimethamine_smiles)
logp_pyr = Descriptors.MolLogP(mol_pyr)
print(f"pyrimethamine LogP: {logp_pyr:.2f}")

# Isoniazid 
isoniazid_smiles = "C1=CN=CC=C1C(=O)NN"
mol_iso= Chem.MolFromSmiles(isoniazid_smiles)
logp_iso = Descriptors.MolLogP(mol_iso)
print(f"Isoniazid  LogP: {logp_iso:.2f}")

```

"We've calculated LogP values. All are under 5. Note Isoniazid and metformin has negative LogP, meaning the two drugs prefers water over fat."
