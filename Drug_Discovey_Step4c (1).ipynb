Hydrogen Bond Acceptors - Second Lipinski Rule Explanation Script: "The second Lipinski rule states hydrogen bond acceptors should be less than 10. Acceptors are typically oxygen and nitrogen atoms. Let's calculate this for our drug set."

```python

from rdkit import Chem
from rdkit.Chem import Lipinski

# Chloroquine 
chloroquine_smiles = "CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl"
mol_chl = Chem.MolFromSmiles(chloroquine_smiles)
hbd_chl = Lipinski.NumHAcceptors(mol_chl)
print(f"chloroquine H-bond acceptors: {hbd_chl}")

# Metformin 
metformin_smiles = "CN(C)C(=N)N=C(N)N"
mol_met = Chem.MolFromSmiles(metformin_smiles)
hbd_met = Lipinski.NumHAcceptors(mol_met)
print(f"Metformin H-bond acceptors: {hbd_met}")

# Ibuprofen
ibuprofen_smiles = "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"
mol_ibu = Chem.MolFromSmiles(ibuprofen_smiles)
hbd_ibu = Lipinski.NumHAcceptors(mol_ibu)
print(f"Ibuprofen H-bond acceptors: {hbd_ibu}")

# Pyrimethamine
pyrimethamine_smiles = "CCC1=C(C(=NC(=N1)N)N)C2=CC=C(C=C2)Cl"
mol_pyr = Chem.MolFromSmiles(pyrimethamine_smiles)
hbd_pyr = Lipinski.NumHAcceptors(mol_pyr)
print(f"pyrimethamine H-bond acceptors: {hbd_pyr}")

# Isoniazid 
isoniazid_smiles = "C1=CN=CC=C1C(=O)NN"
mol_iso = Chem.MolFromSmiles(isoniazid_smiles)
hbd_iso = Lipinski.NumHAcceptors(mol_iso)
print(f"Isoniazid H-bond acceptors: {hbd_iso}")


```

"We've calculated hydrogen bond acceptors. All are under 10, so these drugs pass the second rule as well."
