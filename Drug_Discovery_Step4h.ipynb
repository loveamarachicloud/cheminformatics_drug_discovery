# Step 4h: Checking Lipinski Rule Compliance
Explanation Script: "Having the values is useful, but we need to check which drugs actually pass all four rules. Let's add a pass/fail column to our analysis."

```python
from rdkit import Chem
from rdkit.Chem import Descriptors, Lipinski

drugs  = [ 
("metformin", "CN(C)C(=N)N=C(N)N"),
("chloroquine", "CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl"), 
("isoniazid", "C1=CN=CC=C1C(=O)NN"), 
("pyrimethamine", "CCC1=C(C(=NC(=N1)N)N)C2=CC=C(C=C2)Cl"),
("ibuprofen","CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"),
]
print("Drug        MW    LogP  HBD  HBA  Pass/Fail")
print("-" * 45)

for name, smiles in drugs:
    mol = Chem.MolFromSmiles(smiles)
    mw = Descriptors.MolWt(mol)
    logp = Descriptors.MolLogP(mol)
    hbd = Lipinski.NumHDonors(mol)
    hba = Lipinski.NumHAcceptors(mol)
    
    violations = 0
    if mw >= 500: violations += 1
    if logp >= 5: violations += 1
    if hbd >= 5: violations += 1
    if hba >= 10: violations += 1
    
    status = "PASS" if violations == 0 else f"FAIL ({violations})"
    
    print(f"{name:10} {mw:5.1f} {logp:5.2f} {hbd:4} {hba:4}  {status}")




```

"All five drugs pass Lipinski's rules. This is why they're successful oral medications. In drug discovery, compounds that fail multiple rules are often filtered out early."
