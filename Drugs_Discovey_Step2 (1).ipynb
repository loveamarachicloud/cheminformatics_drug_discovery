# Step 2: What's a Molecule? SMILES Strings Explained
Explanation Script: "Computers don't understand chemical structures directly. Instead, we use SMILES – a text representation. For example, chloroquine's SMILES is CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl. Let's use RDKit to convert that SMILES into a molecule object and visualize it. Run this code in Jupyter:"

```python
from rdkit import Chem
from rdkit.Chem import Draw

# Chloroquine SMILES
aspirin_smiles = "CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl"

# Create molecule object
mol = Chem.MolFromSmiles(aspirin_smiles)

# Draw the molecule (image will appear in notebook)
Draw.MolToImage(mol)
```
