# Step 4f: Topological Polar Surface Area (TPSA)
Explanation Script: "TPSA measures the polar surface area of a molecule. It predicts cell membrane permeability and oral bioavailability. Lower TPSA generally means better absorption."

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

# Chloroquine 
chloroquine_smiles = "CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl"
mol_chloro = Chem.MolFromSmiles(chloroquine_smiles)
tpsa_chloro = Descriptors.TPSA(mol_chloro)
print(f"chloroquine TPSA: {tpsa_chloro:.1f} Å²")

# Metformin 
metformin_smiles = "CN(C)C(=N)N=C(N)N"
mol_met = Chem.MolFromSmiles(metformin_smiles)
tpsa_met = Descriptors.TPSA(mol_met)
print(f"metformin  TPSA: {tpsa_met:.1f} Å²")

# Pyrimethamine
pyrimethamine_smiles = "CCC1=C(C(=NC(=N1)N)N)C2=CC=C(C=C2)Cl"
mol_pyr = Chem.MolFromSmiles(pyrimethamine_smiles)
tpsa_pyr = Descriptors.TPSA(mol_pyr)
print(f"pyrimethamine TPSA: {tpsa_pyr:.1f} Å²")

# Ibuprofen
ibuprofen_smiles = "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"
mol_ibu = Chem.MolFromSmiles(ibuprofen_smiles)
tpsa_ibu = Descriptors.TPSA(mol_ibu)
print(f"Ibuprofen TPSA: {tpsa_ibu:.1f} Å²")

# Isoniazid 
isoniazid_smiles = "C1=CN=CC=C1C(=O)NN"
mol_iso = Chem.MolFromSmiles(isoniazid_smiles)
tpsa_iso = Descriptors.TPSA(mol_iso)
print(f"isoniazidTPSA: {tpsa_iso:.1f} Å²")

```

"We've calculated TPSA values. All are under 140 Å², which is good for oral drugs. chloroquine has lowest TPSA, matching its higher LogP."
