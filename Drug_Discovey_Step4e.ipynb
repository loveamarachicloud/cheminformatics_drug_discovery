# Step 4e: Molecular Weight - Fourth Lipinski Rule
Explanation Script: "The fourth Lipinski rule requires molecular weight under 500. Larger molecules struggle to cross cell membranes. Let's check our drugs."

```python
from rdkit import Chem
from rdkit.Chem import Descriptors

drugs  = [ 
("metformin", "CN(C)C(=N)N=C(N)N"),
("chloroquine", "CCN(CC)CCCC(C)NC1=C2C=CC(=CC2=NC=C1)Cl"), 
("isoniazid", "C1=CN=CC=C1C(=O)NN"), 
("pyrimethamine", "CCC1=C(C(=NC(=N1)N)N)C2=CC=C(C=C2)Cl"),
("ibuprofen","CC(C)CC1=CC=C(C=C1)C(C)C(=O)O"),
]

for name, smiles in drugs:
    mol = Chem.MolFromSmiles(smiles)
    mw = Descriptors.MolWt(mol)
    print(f"{name}: {mw:.1f} g/mol")
```

"We've calculated molecular weights. All are well under 500, confirming these are drug-like molecules."
